# Importer pandas_profiling
Pour importer `pandas_profiling`, faites ce qui suit:

````python
import pandas as pd
from pandas_profiling import ProfileReport
````

## Importer des données
Importons maintenant notre jeu de données habituel:

````python
df = pd.read_csv('salesCerealsOriginal.csv', parse_dates=['WEEK_END_DATE'])
df.head()
````

# Générer et consulter un profil
Nous générons maintenant le rapport de profil.

````python
profile = ProfileReport(df,
                        title='Pandas Profiling Report',
                        html={'style':{'full_width':True}})
````

Nous pouvons accéder à ce rapport de deux manières. La première approche consiste à utiliser des widgets (voir ci-dessous).

````python
profile.to_widgets()
````

La deuxième approche consiste en un rapport HTML (voir ci-dessous).

````python
profile.to_notebook_iframe()
````

Si nécessaire, il est également possible d'enregistrer ce rapport sous forme de rapport HTML ou de données json.
