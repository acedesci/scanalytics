### Exercise 3: Total Costs
Using the unit acquisition cost, quantity ordered, safety stock, annual demand, unit ordering cost, and annual inventory cost of a unit, calculate the total annual cost using the information below. The total annual cost consists of the sum of the annual acquisition cost, the total ordering cost, and the annual holding cost, as follows:

$$TC=PD + S\left(\frac{D}{Q}\right) + H\left(\frac{Q}{2}+ss\right)$$

- $TC$: total annual cost;
- $P$: purchase unit price (=$\$5.6$);
- $D$: annual demand (=$15000$ units);
- $S$: fixed cost per order (=$\$45$);
- $Q$:  order quantity  (=$200$ units);
- $H$: unit inventory holding cost  (=$\$0.8$);
- $ss$: safety stock (=$50$ units).

```python
### start your code here ###

total_annual_cost = 5.6 * 15000 + 45 * (15000/200) + 0.8 * ((200/2) + 50)
### end your code here ####
print('The total annual cost is: ', total_annual_cost)
```
